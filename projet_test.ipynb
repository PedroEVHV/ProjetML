```python
# Exercice 3 (Tp apprentissage supervisé)
import matplotlib.pyplot as plt
# Importer datasets, classifieur (svm) and métriques de performance
from sklearn import datasets, svm, metrics
from sklearn.model_selection import train_test_split
import joblib
from google.colab import files
digits = datasets.load_digits()
_, axes = plt.subplots(nrows=1, ncols=4, figsize=(10, 3))
for ax, image, label in zip(axes, digits.images, digits.target):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title('Training: %i' % label)
# Aplatir (vectoriser) les images
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))
# Diviser les données en 50% pour le l’entraînement et 50% pour le test
X_train, X_test, y_train, y_test = train_test_split(data, digits.target, test_size=0.5, shuffle=False)

# Créer un classificateur de type « Support Vector Machine »
# Gamma : définit à quel point les points éloignés influencent le calcul de la possible ligne de séparation
# Gamma élevé : signifie que seulement les points proches auront une forte influence sur le choix de la ligne de séparation;
# Gamma faible : signifie que les points éloignés doivent également être pris en compte pour obtenir la limite de décision de la ligne de séparation.

# Modèle gamma faible
clf = svm.SVC(gamma=0.001, probability=True)
# Le classificateur créé s’entraîne sur le training set
clf.fit(X_train, y_train)
# Prédire la valeur du digit sur le test set
predicted = clf.predict(X_test)
_, axes = plt.subplots(nrows=1, ncols=4, figsize=(10, 3))
for ax, image, prediction in zip(axes, X_test, predicted):
    ax.set_axis_off()
    image = image.reshape(8, 8)
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title(f'Prediction: {prediction}')
print(f"Classification report for classifier {clf}:\n"
      f"{metrics.classification_report(y_test, predicted)}\n")
#ConfusionMatrixDisplay
cm = metrics.confusion_matrix(y_test, predicted)
disp = metrics.ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
# Sauvegarder le modèle entraîné
joblib.dump(clf, 'svm_digits_model.pkl')
# Télécharger le modèle
files.download('svm_digits_model.pkl')


# Modèle gamma élevée
clf2 = svm.SVC(gamma=0.01, probability=True)
# Le classificateur créé s’entraîne sur le training set
clf2.fit(X_train, y_train)
# Prédire la valeur du digit sur le test set
predicted2 = clf2.predict(X_test)
_, axes2 = plt.subplots(nrows=1, ncols=4, figsize=(10, 3))
for ax, image, prediction in zip(axes2, X_test, predicted2):
    ax.set_axis_off()
    image = image.reshape(8, 8)
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title(f'Prediction: {prediction}')
print(f"Classification report for classifier {clf2}:\n"
      f"{metrics.classification_report(y_test, predicted2)}\n")
#ConfusionMatrixDisplay
cm2 = metrics.confusion_matrix(y_test, predicted2)
disp2 = metrics.ConfusionMatrixDisplay(confusion_matrix=cm2)
disp2.plot()
plt.show()
# Sauvegarder le modèle entraîné
joblib.dump(clf, 'svm_digits_model_2.pkl')
# Télécharger le modèle
files.download('svm_digits_model_2.pkl')


# Modèle gamma plus faible
clf3 = svm.SVC(gamma=0.0001, probability=True)
# Le classificateur créé s’entraîne sur le training set
clf3.fit(X_train, y_train)
# Prédire la valeur du digit sur le test set
predicted3 = clf3.predict(X_test)
_, axes3 = plt.subplots(nrows=1, ncols=4, figsize=(10, 3))
for ax, image, prediction in zip(axes3, X_test, predicted3):
    ax.set_axis_off()
    image = image.reshape(8, 8)
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title(f'Prediction: {prediction}')
print(f"Classification report for classifier {clf3}:\n"
      f"{metrics.classification_report(y_test, predicted3)}\n")
#ConfusionMatrixDisplay
cm3 = metrics.confusion_matrix(y_test, predicted3)
disp3 = metrics.ConfusionMatrixDisplay(confusion_matrix=cm3)
disp3.plot()
plt.show()
# Sauvegarder le modèle entraîné
joblib.dump(clf, 'svm_digits_model_3.pkl')
# Télécharger le modèle
files.download('svm_digits_model_3.pkl')
```
